## Membuat dataset array

```python
X = np.array(df.drop(['high', 'low', 'close'], axis = 1))
y = np.array(df['close'])
```

```python
print(X.shape)
print(y.shape)
```

## Menskalakan data dengan MinMaxScaler

```python
X = MinMaxScaler().fit_transform(X)
```

```python
t = np.reshape(y, (len(y),1))
t = MinMaxScaler().fit_transform(t)
y = t.reshape(-1)
```

```python
print(X.max())
print(X.min())
```

```python
print(y.max())
print(y.min())
```

## Membuat array numpy berbasis deret waktu

```python
length = 120

X_res = []
y_res = []

for i in range(length , len(X)) :
    X_res.append(X[i-length:i])
    y_res.append(y[i])

X_res = np.array(X_res)
y_res = np.array(y_res)
```

```python
print(X_res.shape)
print(y_res.shape)
```

## Prediksi Aluminum

```python
for i, layer in enumerate(model.layers) :
    if i < 5 :
        layer.trainable = True # Layer akan dilatih lagi
```

## Train test split

```python
X_train, X_test, y_train, y_test  = train_test_split(X_res, y_res, train_size = 0.8, test_size = 0.2, random_state = 1)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

## Training Model

```python
checkp = [tf.keras.callbacks.ModelCheckpoint('/content/drive/MyDrive/TransferLearning/LSTM/RMSProp/Model_3/result_metal_3.h5', save_best_only = False, verbose = 1), 
          tf.keras.callbacks.CSVLogger('/content/drive/MyDrive/TransferLearning/LSTM/RMSProp/Model_3/result_metal_3.log', separator=",", append=False)]
```

```python
history = model.fit(
    X_train, 
    y_train, 
    epochs = 200, 
    batch_size = 64, 
    validation_data = (X_test, y_test), 
    callbacks = [checkp])
```

```python
plt.figure(figsize = (16, 8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train_loss', 'validation_loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs Aluminum RMSProp')
plt.savefig('/content/drive/MyDrive/TransferLearning/LSTM/RMSProp/Model_3/result_metal_3_loss.png')
plt.show()
```

```python
plt.figure(figsize = (16, 8))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['accuracy', 'validation_accuracy'])
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.title('accuracy vs Epochs Aluminum RMSProp')
plt.savefig('/content/drive/MyDrive/TransferLearning/LSTM/RMSProp/Model_3/result_metal_3_acc.png')
plt.show()
```

## Prediksi

```python
model = load_model('/content/drive/MyDrive/TransferLearning/LSTM/RMSProp/Model_3/result_metal_3.h5')
pred = model.predict(X_test)
```

```python
print(pred.shape)
```

## Plot the Prediction and Test value

```python
plt.figure(figsize = (20,8))
plt.plot(y_test)
plt.plot(pred)
plt.xlabel('Periode')
plt.ylabel('Skala')
plt.legend(['testing price','predicted price'])
plt.title('Perbandingan skala harga test dan prediksi close untuk Aluminum 78 Periode menggunakan RMSProp', fontsize = 15)
plt.savefig('/content/drive/MyDrive/TransferLearning/LSTM/RMSProp/Model_3/result_metal_3_comparison_testingVSpredic_78_periode.png')
```

## Performance Metrics

```python
print('r2 score : ' + str(r2_score(y_test, pred)))
print('Mean Squared Error (MSE) : ' + str(mean_squared_error(y_test, pred)))
print('Root Mean Squared Error (RMSE) : ' + str(mean_squared_error(y_test, pred, squared=False)))
print('Mean Absolut Error (MAE) : ' + str(mean_absolute_error(y_test, pred)))
print('Mean Absolut Percentage Error (MAPE) : ' + str(mean_absolute_percentage_error(y_test, pred)))
```
